```python
# Q4. What are the main categories of anomaly detection algorithms?
```

```python
# 1. Statistical-Based Methods
# These methods assume a statistical model for the data and identify anomalies as data points that are unlikely to be 
# generated by the model.
# Examples: Z-score method, Modified Z-score method, Chi-square method

# 2. Machine Learning-Based Methods
# These methods use machine learning algorithms to learn the patterns in the data and identify anomalies.
# Examples: One-Class SVM, Local Outlier Factor (LOF), Isolation Forest

# 3. Distance-Based Methods
# These methods calculate the distance between data points and identify anomalies as points that are farthest from the rest.
# Examples: k-Nearest Neighbors (k-NN), k-Means

# 7. Graph-Based Methods
# These methods are used for anomaly detection in graph-structured data, such as social networks or web graphs.
# Examples: Graph-Based Anomaly Detection (GBAD
```

```python
# Q5. What are the main assumptions made by distance-based anomaly detection methods?
```

```python
# Normal data is dense and close together
# Anomalies are sparse, isolated, and far from normal data
# The chosen distance metric is meaningful
# Data is uniformly distributed
# Anomalies are not clustered
# Normal data is not noisy
# A clear distance threshold can be defined
```

```python
# Q6. How does the LOF algorithm compute anomaly scores?
```

```python
# k-Nearest Neighbors: Identify the k nearest neighbors of each data point.
# Reachability Distance: Measure the reachability distance to avoid outliers having too much influence.
# Local Reachability Density (LRD): Calculate the local density of each point based on reachability distances.
# LOF Score: Compare the LRD of a point to its neighbors to compute the LOF score.

# Interpretation:
# LOF â‰ˆ 1: Normal point with density similar to neighbors.
# LOF > 1: Potential outlier with lower density than neighbors.
# LOF < 1: Less likely to be an outlier, in a denser region.
```
