```python
# Q2: What is a scatter plot? Use the following code to generate data for x and y. Using this generated data
# plot a scatter plot.
# import numpy as np
# np.random.seed(3)
# x = 3 + np.random.normal(0, 2, 50)
# y = 3 + np.random.normal(0, 2, len(x))
# Note: Also add title, xlabel, and ylabel to the plot.
```

```python
 # Ans: A scatter plot displays values for two variables as a collection of points on a graph, with one variable on the
 #    x-axis and the other on the y-axis. It is useful for visualizing the relationship and identifying potential
 #    correlations or patterns between two continuous variables.
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
```

```python
plt.scatter(x, y, color='blue', edgecolor='k', alpha=0.7)
plt.title('Scatter Plot of Generated Data')
plt.xlabel('X values')
plt.ylabel('Y values')
plt.show()
```

```python
# Q3: Why is the subplot() function used? Draw four line plots using the subplot() function.
# Use the following data:
# import numpy as np
# For line 1: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([0, 100, 200, 300, 400, 500])
# For line 2: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([50, 20, 40, 20, 60, 70])
# For line 3: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([10, 20, 30, 40, 50, 60])
# For line 4: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([200, 350, 250, 550, 450, 150])
```

```python
# Ans: The subplot() function in Matplotlib is used to create multiple plots within a single figure, allowing for the
# organization and arrangement of plots in a grid layout. This facilitates the comparison and contrast of different 
# visualizations.
```

```python
x1 = np.array([0, 1, 2, 3, 4, 5])
y1 = np.array([0, 100, 200, 300, 400, 500])

x2 = np.array([0, 1, 2, 3, 4, 5])
y2 = np.array([50, 20, 40, 20, 60, 70])

x3 = np.array([0, 1, 2, 3, 4, 5])
y3 = np.array([10, 20, 30, 40, 50, 60])

x4 = np.array([0, 1, 2, 3, 4, 5])
y4 = np.array([200, 350, 250, 550, 450, 150])
```

```python
fig, axs = plt.subplots(2, 2, figsize=(10, 8))
axs[0, 0].plot(x1, y1, color='b')
axs[0, 0].set_title('Line 1')
axs[0, 0].set_xlabel('x')
axs[0, 0].set_ylabel('y')

axs[0, 1].plot(x2, y2, color='g')
axs[0, 1].set_title('Line 2')
axs[0, 1].set_xlabel('x')
axs[0, 1].set_ylabel('y')

axs[1, 0].plot(x3, y3, color='r')
axs[1, 0].set_title('Line 3')
axs[1, 0].set_xlabel('x')
axs[1, 0].set_ylabel('y')

axs[1, 1].plot(x4, y4, color='m')
axs[1, 1].set_title('Line 4')
axs[1, 1].set_xlabel('x')
axs[1, 1].set_ylabel('y')

plt.show()
```

```python
# Q4: What is a bar plot? Why is it used? Using the following data plot a bar plot and a horizontal bar plot.
# import numpy as np
# company = np.array(["Apple", "Microsoft", "Google", "AMD"])
# profit = np.array([3000, 8000, 1000, 10000])
```

```python
# Ans: A bar plot is a chart that presents categorical data with rectangular bars. Each bar's height or length is
# proportional to the value it represents. Bar plots are used to:

# Compare Values: Easily compare quantities across different categories.
# Highlight Differences: Visually highlight differences between different groups.
# Show Trends: Display changes or trends in data over time (when categories represent time periods).
# Summarize Data: Provide a clear summary of large sets of data in a simple visual format.
```

```python
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])
```

```python
plt.subplot(1, 2, 1)
plt.bar(company, profit, color='skyblue')
plt.title('Company Profits')
plt.xlabel('Company')
plt.ylabel('Profit (in millions)')
plt.show()
```

```python
plt.subplot(1, 2, 2)
plt.barh(company, profit, color='lightgreen')
plt.title('Company Profits')
plt.xlabel('Profit (in millions)')
plt.ylabel('Company')
plt.show()
```

```python
# Q5: What is a box plot? Why is it used? Using the following data plot a box plot.
# box1 = np.random.normal(100, 10, 200)
# box2 = np.random.normal(90, 20, 200)
```

```python
# Ans: A box plot, or whisker plot, visually represents data distribution using a five-number summary: minimum,
# first quartile (Q1), median (Q2), third quartile (Q3), and maximum. It highlights quartiles and potential outliers,
# making it useful for comparing different data sets.

# Why is it Used?
# Visualizing Distribution: Highlights central tendency and variability.
# Identifying Outliers: Detects data points significantly outside the normal range.
# Comparing Distributions: Compares multiple data sets side by side.
# Understanding Spread and Skewness: Provides insights into data spread and skewness.
```

```python
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)
```

```python
plt.boxplot([box1, box2], labels=['Box1', 'Box2'])
plt.title('Box Plot of Two Distributions')
plt.xlabel('Box')
plt.ylabel('Value')
plt.show()
```
