# Que 1: Name any five plots that we can plot using the Seaborn library. Also, state the uses of each plot.

```python
# Ans:

# Scatter Plot:
# Use: Visualizes the relationship between two continuous variables, identifying trends, correlations, and outliers.
# Bar Plot:

# Bar Plot:
# Use: Displays the distribution of categorical data, comparing values of different categories to show central tendency and spread.
# Box Plot:

# Box Plot:
# Use: Shows data distribution based on the five-number summary (minimum, Q1, median, Q3, maximum) to identify outliers and compare distributions.
# Histogram:

# Histogram:
# Use: Visualizes the distribution of a single continuous variable by dividing data into bins, helping understand frequency distribution and data shape.
# Heatmap:

# Heatmap:
# Use: Displays the magnitude of data values across a 2D matrix, commonly used for showing correlation matrices and relationships between variables.
    
```

```python
fmri = sns.load_dataset("fmri")
```

```python
sns.lineplot(data=fmri, x="timepoint", y="signal", hue="event", style="region")
```
